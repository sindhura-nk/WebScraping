## Fetch the html code from website

```python
import requests
```

```python
url = "https://www.thehindu.com/latest-news/"
response = requests.get(url)
response
```

```python
data = response.content
print(str(data)[0:1000])
```

```python
from bs4 import BeautifulSoup
bs = BeautifulSoup(data)
```

```python
## Finding news articles
news = bs.find_all('div',class_='right-content')
news
```

```python
news[0]
```

```python
news[0].text
```

```python
news_titles = []
for i in news:
    news_titles.append(i.text)
news_titles
```

```python
latest_date = bs.find_all('div',class_='latest-date')
latest_date[0].text
```

```python
latest_date[0].text.split(',')
```

```python
day = latest_date[0].text.split(',')[0]
date = latest_date[0].text.split(',')[1]
```

```python
day
```

```python
date
```

```python
date =date.split('\n')[0]
date
```

```python
news[0].a.get('href')
```

```python
links = []
for i in news:
    links.append(i.a.get('href'))
links
```

## Create a dataframe containing all values

```python
dct = {'News Titles':news_titles,
      'Day':day,
      'Date':date,
       'Links':links
      }
```

```python
import pandas as pd
```

```python
df_news = pd.DataFrame(dct)
df_news
```

## Saving above to csv

```python
df_news.to_csv('Webscaping News Articles.csv',index=False)
```
